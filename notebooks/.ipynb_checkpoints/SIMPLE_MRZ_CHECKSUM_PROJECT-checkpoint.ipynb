## Part 1: Understanding MRZ (Machine Readable Zone)

### What is MRZ?
The bottom part of passport with numbers and letters.

### Example Passport Data:
```
Passport Number: L898902C3
Country: IND
Date of Birth: 01-02-1980
Gender: M
Expiry Date: 23-12-2010
```

### MRZ Format:
```
L898902C3<6IND800201M1012230<<<<<<<<<<<<<<<0
```

### Breaking it down:
- `L898902C3` - Passport number
- `<` - Filler  
- `6` - Checksum for passport number
- `IND` - Country code
- `800201` - Date of birth (YYMMDD)
- `M` - Gender
- `101223` - Expiry date (YYMMDD)
- `0` - Final checksum

---

## Part 2: MRZ Checksum Calculation

### Rules (Simple!):

1. **Convert characters to numbers:**
   - Digits (0-9) â†’ Same value
   - Letters (A-Z) â†’ A=10, B=11, C=12, ..., Z=35
   - '<' symbol â†’ 0

2. **Multiply by weights:**
   - Weight pattern: 7, 3, 1, 7, 3, 1, 7, 3, 1... (repeating)

3. **Calculate:**
   - Sum all (value Ã— weight)
   - Take mod 10 (remainder when divided by 10)
   - Result = Checksum

---

````python
# Function to convert character to number
def char_to_number(char):
    """
    Convert MRZ character to number.
    
    Examples:
    - '5' â†’ 5
    - 'A' â†’ 10
    - 'L' â†’ 21
    - '<' â†’ 0
    """
    
    # If it's a digit (0-9)
    if char.isdigit():
        return int(char)
    
    # If it's a letter (A-Z)
    elif char.isalpha():
        # A=10, B=11, ..., Z=35
        return ord(char.upper()) - ord('A') + 10
    
    # If it's '<' or any other character
    else:
        return 0

# Test the function
print("Testing char_to_number:")
print(f"'5' â†’ {char_to_number('5')}")
print(f"'A' â†’ {char_to_number('A')}")
print(f"'L' â†’ {char_to_number('L')}")
print(f"'<' â†’ {char_to_number('<')}")
````

````python
# Function to calculate checksum
def calculate_checksum(data):
    """
    Calculate MRZ checksum using mod 10 algorithm.
    
    Steps:
    1. Convert each character to number
    2. Multiply by weight (7, 3, 1, repeating)
    3. Add all values
    4. Take mod 10
    
    Example:
    data = 'L898902C3'
    L(21)Ã—7 + 8Ã—3 + 9Ã—1 + 8Ã—7 + 9Ã—3 + 0Ã—1 + 2Ã—7 + C(12)Ã—3 + 3Ã—1
    = 147 + 24 + 9 + 56 + 27 + 0 + 14 + 36 + 3
    = 316
    316 % 10 = 6
    """
    
    # Weight pattern (repeating)
    weights = [7, 3, 1]
    
    # Total sum
    total = 0
    
    # Go through each character
    for i, char in enumerate(data):
        # Convert character to number
        value = char_to_number(char)
        
        # Get weight (cycles through 7, 3, 1)
        weight = weights[i % 3]  # % 3 makes it repeat
        
        # Add to total
        total += value * weight
        
        # Debug print (you can remove this later)
        print(f"Position {i}: '{char}' â†’ value={value}, weight={weight}, product={value*weight}")
    
    # Calculate checksum (mod 10)
    checksum = total % 10
    
    print(f"\nTotal sum = {total}")
    print(f"Checksum = {total} % 10 = {checksum}")
    
    return checksum

# Test with passport number
print("Testing checksum calculation:")
print("="*60)
passport_number = 'L898902C3'
calculated = calculate_checksum(passport_number)
expected = 6

print(f"\nPassport Number: {passport_number}")
print(f"Calculated Checksum: {calculated}")
print(f"Expected Checksum: {expected}")
print(f"Match: {calculated == expected}")
````

## Part 3: Verify Complete MRZ

Now we check the whole MRZ line.

````python
# Function to verify MRZ
def verify_mrz(mrz_line):
    """
    Verify MRZ checksum.
    
    MRZ Format:
    L898902C3<6IND800201M1012230<<<<<<<<<<<<<<<0
    
    Positions:
    0-8: Passport number (L898902C3)
    9: Passport checksum (6)
    10-12: Country (IND)
    13-18: Date of birth (800201)
    19: Gender (M)
    20-25: Expiry date (101223)
    43: Final checksum (0)
    """
    
    print("MRZ Verification")
    print("="*60)
    print(f"MRZ Line: {mrz_line}")
    print("="*60)
    
    # Extract passport number and checksum
    passport_number = mrz_line[0:9]
    passport_checksum = int(mrz_line[9]) if mrz_line[9].isdigit() else 0
    
    print(f"\n1. Passport Number: {passport_number}")
    print(f"   Printed Checksum: {passport_checksum}")
    
    # Calculate checksum
    calculated_checksum = calculate_checksum(passport_number)
    
    print(f"   Calculated Checksum: {calculated_checksum}")
    
    # Check if match
    if calculated_checksum == passport_checksum:
        print("   âœ“ MATCH - Passport number is valid")
        result = True
    else:
        print("   âœ— MISMATCH - Passport number is FAKE!")
        result = False
    
    print("="*60)
    
    return result

# Test with real MRZ
print("Testing with REAL passport MRZ:")
print()
real_mrz = 'L898902C3<6IND800201M1012230<<<<<<<<<<<<<<<0'
is_valid = verify_mrz(real_mrz)

print(f"\nFINAL RESULT: {'REAL PASSPORT' if is_valid else 'FAKE PASSPORT'}")
````

````python
# Test with FAKE MRZ (wrong checksum)
print("Testing with FAKE passport MRZ (tampered):")
print()

# Fake: Changed passport number but didn't update checksum
fake_mrz = 'L898902C9<6IND800201M1012230<<<<<<<<<<<<<<<0'  # Changed C3 to C9, but checksum still 6
is_valid = verify_mrz(fake_mrz)

print(f"\nFINAL RESULT: {'REAL PASSPORT' if is_valid else 'FAKE PASSPORT'}")
````

## Part 4: Optional CNN Check (May Fail)

CNN looks at the image but it's NOT the final decision.
MRZ checksum is always the final authority.

````python
# Simple CNN check function
def cnn_visual_check(image_path):
    """
    Optional: Check passport image using CNN.
    
    Returns:
    - confidence: Number between 0 and 1
    - If < 0.5: CNN unclear, move to MRZ
    - If >= 0.5: CNN thinks real, but still check MRZ!
    """
    
    print("Stage 1: CNN Visual Check")
    print("="*60)
    
    try:
        # Load image
        img = cv2.imread(image_path)
        
        if img is None:
            print("ERROR: Could not load image")
            return 0.0
        
        # Resize to 224x224 for CNN
        img = cv2.resize(img, (224, 224))
        
        # Normalize (0-1)
        img = img / 255.0
        
        # Note: In real project, you'd load trained model here
        # For demo, we'll use random score
        print("âš  Model not trained. Using placeholder score.")
        confidence = 0.45  # Simulated low confidence
        
        print(f"CNN Confidence: {confidence:.2f}")
        
        if confidence < 0.5:
            print("âŒ CNN Result: UNCLEAR (confidence too low)")
            print("â†’ Moving to MRZ verification (final check)")
        else:
            print("âœ“ CNN Result: Looks real")
            print("â†’ But still need MRZ verification!")
        
        print("="*60)
        return confidence
        
    except Exception as e:
        print(f"ERROR in CNN: {e}")
        print("â†’ Moving to MRZ verification")
        print("="*60)
        return 0.0
````

## Part 5: Complete Two-Stage Verification System

````python
def verify_passport(image_path, mrz_line):
    """
    Complete two-stage passport verification.
    
    Stage 1: CNN visual check (optional, may fail)
    Stage 2: MRZ checksum (final decision)
    
    Parameters:
    - image_path: Path to passport image
    - mrz_line: MRZ text from passport
    
    Returns:
    - Final decision: REAL or FAKE
    """
    
    print("\n" + "#"*60)
    print("#" + " "*58 + "#")
    print("#" + "  PASSPORT VERIFICATION SYSTEM".center(58) + "#")
    print("#" + " "*58 + "#")
    print("#"*60 + "\n")
    
    # STAGE 1: CNN CHECK (Optional)
    print("\nðŸ“· STAGE 1: CNN VISUAL CHECK (Optional)\n")
    
    if image_path:
        cnn_confidence = cnn_visual_check(image_path)
    else:
        print("No image provided. Skipping CNN check.")
        print("="*60)
        cnn_confidence = 0.0
    
    print("\nâ†’ CNN result is NOT final. Proceeding to MRZ...\n")
    
    # STAGE 2: MRZ CHECKSUM (Final Decision)
    print("\nðŸ”¢ STAGE 2: MRZ CHECKSUM VALIDATION (Final Decision)\n")
    
    mrz_valid = verify_mrz(mrz_line)
    
    # FINAL DECISION (based ONLY on MRZ)
    print("\n" + "="*60)
    print("FINAL DECISION")
    print("="*60)
    
    if mrz_valid:
        print("\nâœ“âœ“âœ“ PASSPORT IS REAL âœ“âœ“âœ“")
        print("Reason: MRZ checksum matches")
        result = "REAL"
    else:
        print("\nâœ—âœ—âœ— PASSPORT IS FAKE âœ—âœ—âœ—")
        print("Reason: MRZ checksum mismatch")
        result = "FAKE"
    
    print("="*60 + "\n")
    
    return result
````
